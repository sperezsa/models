```python
# Cargar el modelo de español
nlp = spacy.load("es_core_news_sm")

# Procesar un texto
doc = nlp("Hola, mundo. Este es un ejemplo de procesamiento de lenguaje natural con spaCy.")

# Iterar sobre los tokens
for token in doc:
    print(token.text, token.pos_)
```

```python
# Encontrar entidades nombradas
for ent in doc.ents:
    print(ent.text, ent.label_)
```
