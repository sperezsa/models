```python
from transformers import ViTImageProcessor, ViTForImageClassification
from PIL import Image
import requests
```

```python
# Option 1: Load image from Google Drive
from google.colab import drive
drive.mount('/content/drive')
image = Image.open('/content/drive/MyDrive/tu_ruta/1366_2000.jpg')
#image = Image.open('/content/drive/MyDrive/tu_ruta/foto-gato-siames-cachorro-768x402.jpg')
#image = Image.open('/content/drive/MyDrive/tu_ruta/IMG-2847-scaled.jpg')


# Another option by URL
#url = 'http://images.cocodataset.org/val2017/000000039769.jpg'
#image = Image.open(requests.get(url, stream=True).raw)
```

```python
processor = ViTImageProcessor.from_pretrained('google/vit-base-patch16-224')
model = ViTForImageClassification.from_pretrained('google/vit-base-patch16-224')
```

```python
inputs = processor(images=image, return_tensors="pt")
outputs = model(**inputs)
logits = outputs.logits
# model predicts one of the 1000 ImageNet classes
predicted_class_idx = logits.argmax(-1).item()
print("Predicted class:", model.config.id2label[predicted_class_idx])
```

```python
# To show the top 5 predictions
import torch

# Get the top 5 predictions
top_k_probs, top_k_indices = torch.topk(logits, k=5, dim=1)

for i in range(5):
    predicted_class_idx = top_k_indices[0][i].item()
    probability = top_k_probs[0][i].item()
    print(f"Predicted class {i+1}: {model.config.id2label[predicted_class_idx]} with probability {probability:.2f}")

print("top_k_probs", top_k_probs)
print("top_k_indices", top_k_indices)
```
